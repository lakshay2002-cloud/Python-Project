```python
import pandas as pd 
import matplotlib.pyplot as plt 

data = pd.read_csv(r"C:\Users\Lenovo\OneDrive\Documents\Projects\Luke Barousse Projects\Python Course  Data Analysis\data_jobs.csv")
df=data
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
```

```python
df['job_posted_month']= df['job_posted_date'].dt.strftime('%b')
```

```python
df
```

```python
months = df['job_posted_month'].unique()
{month :month for month in months}
```

```python
dict_month={month : df[df['job_posted_month'] == month ]for month in months}
```

```python
dict_month['Jan']
```
